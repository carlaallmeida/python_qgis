# O que é um coregistro de imagens?

O coregistro de imagens é um processo de georreferenciamento que sobrepõe e alinha duas ou mais imagens, utilizando uma imagem de referência e uma imagem alvo. O coregistro geoespacial é um pré-requisito essencial no processamento de dados de sensoriamento remoto para garantir a qualidade e confiabilidade das análises, visto que desalinhamentos entre sensores podem comprometer severamente as análises de imagens subsequentes.

O coregistro pode ser realizado de forma **intrasensor**, quando as imagens pertencem ao mesmo sensor em diferentes datas de aquisição, ou **intersensor**, quando as imagens pertencem a sensores diferentes.

- No coregistro intrasensor, o objetivo principal é alinhar imagens multitemporais para análises de mudanças, séries temporais e monitoramento ambiental. O processo tende a ser mais simples, pois as imagens geralmente possuem características semelhantes de resolução espacial, geometria e resposta espectral.
- O coregistro intersensor, o objetivo é alinhar espacialmente dados de diferentes sensores, reduzindo diferenças temporais da superfície e aumentando a precisão geométrica para análises multissensores e multitemporais. Esse, envolve o alinhamento de imagens com diferentes resoluções espaciais, geometrias, bandas espectrais e sistemas de aquisição. Esse tipo de coregistro é mais desafiador devido às diferenças radiométricas e geométricas entre os sensores.

# Objetivo

Neste projeto, será aplicada a técnica de coregistro utilizando o pacote AROSICS. Como imagem alvo, será utilizada uma imagem do sensor MUX CBERS-4A em nível de processamento L2; e como imagem referência, correspondente do Sentinel-2 em nível de processamento L1C.

# Metodologia

As imagens utilizadas neste estudo foram obtidas a partir do catálogo de imagens do Instituto Nacional de Pesquisas Espaciais (INPE) e da plataforma Copernicus Data Space Ecosystem. O objetivo da seleção foi garantir a compatibilidade geométrica e espectral entre os sensores, minimizando diferenças temporais e radiométricas que pudessem comprometer o processo de coregistro.

Foi selecionada uma imagem do sensor MUX, embarcado no satélite CBERS-4A, disponível no catálogo do INPE em nível de processamento L2, utilizada como imagem alvo (target image). Esse nível de processamento inclui correção radiométrica e correção geométrica de sistema baseada em dados de efemérides e atitude do satélite, porém ainda não passou por um processo de correção mais refinado, podendo apresentar deslocamentos geométricos residuais.

Como imagem de referência (reference image), foi utilizada uma cena do sensor MSI a bordo do satélite Sentinel-2, obtida no produto SENTINEL-2 L1C disponível no Copernicus Data Space Ecosystem, utilizando o composto LEAST_CC, que prioriza cenas com menor cobertura de nuvens. As imagens Sentinel-2 L1C correspondem a produtos ortorretificados em reflectância no topo da atmosfera (TOA), corrigidos geometricamente com auxílio de Modelo Digital de Elevação (MDE), além de ter máscaras de qualidade e informações atmosféricas auxiliares.

A escolha das imagens considerou os seguintes critérios:

* aquisição em datas próximas, reduzindo diferenças sazonais e alterações na superfície;
* compatibilidade espectral entre as bandas do sensor MUX e do MSI Sentinel-2;

* baixa cobertura de nuvens (máximo 20%).

**Metodologia de coregistro:**

O coregistro geométrico foi realizado utilizando o pacote AROSICS, desenvolvido para correção automática de deslocamentos espaciais entre imagens multitemporais e intersensores. O método implementado pelo AROSICS baseia-se em técnicas de correlação por janela local, permitindo identificar deslocamentos sistemáticos e locais entre a imagem alvo e a imagem de referência

Para aplicar o AROSICS, as imagens precisam estar no formato UTM e coodernadas geográficas. Por isso, como primeira etapa é importante certificar de que as imagens de entrada tenha a mesma projeção. O AROSICS também pode ser aplicado a imagens sem qualquer informação de projeção e geocodificação. Neste caso, as imagens de entrada precisam ter o mesmo tamanho de pixel e devem cobrir mais ou menos a mesma área espacial (com uma mudança de alguns pixels no máximo).

Neste estudo, foi utilizada a abordagem de coregistro local disponível no pacote AROSICS, por apresentar maior precisão no alinhamento geométrico entre imagens multissensores, permitindo corrigir deformações geométricas residuais presentes nas imagens MUX L2 em relação às imagens Sentinel-2 utilizadas como referência.


**Definição das bandas equivalentes entre os sensores:**

Para garantir a compatibilidade espectral entre os sensores, foram selecionadas bandas equivalentes do Sentinel-2 MSI correspondentes às bandas multiespectrais do sensor MUX. A equivalência foi definida com base na proximidade das faixas espectrais e na similaridade das respostas radiométricas observadas em diferentes tipos de superfície.

Foram  utilizadas as bandas B4 (vermelho) e B8 (infravermelho próximo) do Sentinel-2 como correspondentes às bandas 7 e 8 do sensor MUX, respectivamente. Essa compatibilidade espectral favorece a identificação automática de feições homólogas durante o processo de coregistro geométrico.

MUX Banda 7 (Vermelho; 0,63 - 0,69 µm) ↔ Sentinel-2 Banda 4

MUX Banda 8 (Infravermelho próximo; 0,77 - 0,89 µm) ↔ Sentinel-2 Banda 8

# 1. Leitura e preparação de imagens




**1.1 Verificando e carregando a imagem MUX CBERS-4A:**


```python
import os
from qgis.core import QgsProject, QgsRasterLayer

# Caminho da pasta
pasta = r"C:\Users\Tayna\OneDrive\Desktop\Projeto-coregistro\Projeto\CBERS4A"

# Verifica se a pasta existe
if not os.path.exists(pasta):
    print("Pasta não encontrada!")
    print(pasta)

else:
    print("Pasta encontrada!\n")

    arquivos = os.listdir(pasta)

    print(f"Total de arquivos encontrados: {len(arquivos)}\n")

    for f in arquivos:

        print(f"Verificando arquivo: {f}")

        # Filtra apenas TIFF
        if f.lower().endswith((".tif", ".tiff")):

            caminho_completo = os.path.join(pasta, f)

            print(f"Tentando carregar: {caminho_completo}")

            camada = QgsRasterLayer(caminho_completo, f)

            # Verifica se a camada é válida
            if camada.isValid():

                QgsProject.instance().addMapLayer(camada)

                print(f"Camada carregada com sucesso: {f}\n")

            else:
                print(f"ERRO ao carregar: {f}\n")
```

**1.2 Verificando e carregando a imagem Sentinel 2:**

```python
import os
from qgis.core import QgsProject, QgsRasterLayer

# Caminho da pasta com as imagens Sentinel-2
pasta_copernicus = r"C:\Users\Tayna\OneDrive\Desktop\Projeto-coregistro\Projeto\Copernicus"

# Percorre a pasta e adiciona os rasters válidos ao QGIS
for f in os.listdir(pasta_copernicus):

    if f.lower().endswith(('.tif', '.jp2')):

        caminho = os.path.join(pasta_copernicus, f)

        camada = QgsRasterLayer(caminho, f)

        if camada.isValid():
            QgsProject.instance().addMapLayer(camada)
            print(f"Camada Sentinel carregada com sucesso: {f}")

        else:
            print(f"Erro ao carregar: {f}")
```

**1.3 Verificando as projeções das imagens:**

```python
from qgis.core import QgsProject, QgsMapLayerType

# Percorre as camadas raster do projeto para validar o SRC
for camada in QgsProject.instance().mapLayers().values():
    if camada.type() == QgsMapLayerType.RasterLayer:
        crs = camada.crs()

        # Identifica se é Geográfico (Graus) ou Projetado (Metros/UTM)
        tipo = "Geográfico (Graus)" if crs.isGeographic() else "Projetado/UTM (Metros)"

        # Exibe o resultado formatado na tela
        print(f"Imagem: {camada.name()}")
        print(f"  > Código EPSG: {crs.authid()}")
        print(f"  > Nome do SRC: {crs.description()}")
        print(f"  > Unidade:     {tipo}")
        print("-" * 30)
```
