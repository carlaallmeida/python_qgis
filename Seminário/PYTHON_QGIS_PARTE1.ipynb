# **4. Alguns comando importantes**

**4.1 Carregando um shapefile**

```python

#Defina o caminho do shapefile
caminho =r'C:\Users\Tayna\OneDrive\Desktop\AL_Municipios_2025\AL_Municipios_2025.shp'

#Carrega a camada vetorial
camada = QgsVectorLayer(caminho, "AL_Municipios_2025", "ogr")
```



**4.2 Carregando o projeto atual**


*QgsProject.instance()* acessa o projeto atualmente aberto no QGIS.

*.addMapLayer()* Adiciona a camada ao projeto e exibe no mapa

```python
if camada.isValid():
    QgsProject.instance().addMapLayer(camada)
    print("Shapefile carregado!")
else:
    print("Erro ao carregar")
```

**4.3 Listando camadas carregadas**

*project.mapLayers().values()* percorre todas as camadas carregadas no projeto.

.maplayers → Retorna um dicionário com todas as camadas ({id: layer})

.values → Pega apenas os objetos de camada (sem os IDs)

```python
project = QgsProject.instance()
camadas = project.mapLayers().values()

for layer in camadas:
    print(layer.name())
```

**4.9 Filtrar atributos**


```python
#Filtrar municípios com área maior que 500 km2

expression = 'AREA_KM2 > 500'

request = QgsFeatureRequest().setFilterExpression(expression)

for feature in layer.getFeatures(request):
    print(feature['NM_MUN'], feature['AREA_KM2'])
```

**4.10 Criar Seleção automática**

```python
layer.selectByExpression(
    'AREA_KM2 > 500'
)
```

**4.11 Alterar a simbologia via Python**
