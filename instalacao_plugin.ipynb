# <span style="color:#336699">Configuração do Plugin no Qgis(Recomendado)</span>
<hr style="border:1px solid #0077b9;">

Para instalar o plugin, primeiro precisamos adicionar o repositório oficial do OpenGeos seguindo estas etapas:

1. Acesse o menu Complementos -> Gerenciar e Instalar Complementos...

2. No menu lateral esquerdo, clique na aba Configurações.

3. Na seção "Repositórios de Complementos", clique no botão Adicionar.
 
4. Na tela de configuração:

**Nome(Digite):** OpenGeos

**URL(Cole o Link):** [https://qgis.gishub.org/plugins.xml](https://qgis.gishub.org/plugins.xml)

5. Clique em OK (os demais campos podem ficar em branco).

</div>

Dica: Após clicar em **OK**, certifique-se de que o novo repositório aparece como "conectado" na lista para que os plugins fiquem disponíveis para instalação como mostra a imagem abaixo.

### <span style="color:#336699">Instalando o  Plugin</span>
<hr style="border:1px solid #0077b9;">

Após configurar o repositório, o próximo passo é realizar a instalação propriamente dita:

1. Na mesma janela de Complementos, clique na aba Tudo (localizada no menu lateral esquerdo).

2. No campo de busca (pesquisa), digite Notebook.

3. Assim que o plugin aparecer na lista, selecione-o e clique no botão Instalar complemento.

4. Aguarde o término do processo. Uma vez instalado, o ícone do Jupyter Notebook <img src="imagens/iconenotebook.png" width="30" style="vertical-align: middle;"/> aparecerá na sua barra de ferramentas ao lado esquerdo.

<div align="center">
  <img src="imagens/configuracao_parte3.png" width="600"/>
</div>

### <span style="color:#336699">Abrindo o Plugin</span>
<hr style="border:1px solid #0077b9;">

Para acessar o notebook dentro do Qgis clique no icone do notebook <img src="imagens/iconenotebook.png" width="30" style="vertical-align: middle;"/>

<div align="center">
  <img src="imagens/configuracao_parte4.png" width="600"/>
</div>

Dica: Geralmente, ao clicar no ícone , a janela do notebook pode parecer que não abriu ou a imagem pode não aparecer de imediato. Isso acontece porque ela costuma abrir minimizada na parte inferior do QGIS. Para visualizar, basta clicar na borda inferior da tela e arrastar para cima para expandir o painel do Jupyter
